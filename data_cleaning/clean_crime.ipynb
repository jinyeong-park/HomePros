```python
path = Path("C:/Users/gargk/Workspace/5500_EDA/crime_data")

xls_files = Path(path).glob("*.xls")

df_crime = list()
for f in xls_files:
    if len(df_crime) == 0:
        data = pd.read_excel(f, header=3)
        data.columns = ['State', 'City', 'Population', 'Violent Crime',
                        'Murder', 'Rape (revised)', 'Rape (legacy)', 'Robbery',
                        'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
                        'Motor vehicle theft', 'Arson']
        data['file'] = f.stem
        df_crime.append(data)
        print(f.stem)
        print(data.info())
    else:
        data = pd.read_excel(f, header=3)
        data = data.iloc[:, :14]
        # print(data.info())
        print(f.stem)
        data.columns = ['State', 'City', 'Population', 'Violent Crime',
                        'Murder', 'Rape (revised)', 'Rape (legacy)', 'Robbery',
                        'Aggravated Assault', 'Property Crime', 'Burglary', 'Larceny-theft',
                        'Motor vehicle theft', 'Arson']
        data['file'] = f.stem
        print(f.stem)
        print(data.info())
        df_crime.append(data)
        
        

df_crime_concat = pd.concat(df_crime, ignore_index=True)
df_crime_concat
```

```python
df_crime_concat = df_crime_concat.dropna(subset=['City'])
df_crime_concat = df_crime_concat.dropna(axis=0, how='all')
df_crime_concat = df_crime_concat.dropna(subset=['Violent Crime','Property Crime'], how='all')
# df_crime_concat = df_crime_concat[["State", "City", "file", "Population", 'Violent\ncrime', 'Property\ncrime']]
# df_crime_concat.to_csv("crime_concat.csv")
```

```python
def get_year(name):
    match = re.search(r"_(\d\d\d\d)", name)
    if match:
        return match.group(1)
    else:
        return name

df_crime_concat['year'] = df_crime_concat['file'].apply(get_year)
df_crime_concat['year']
```

```python
df_crime_concat['State'].ffill(inplace=True)
df_crime_concat['state'] = df_crime_concat['State'].str.lower().str.capitalize()
df_crime_concat
```

```python
def clean_crime_city(city):
    match = re.search(r"(.*?)\d", city)
    if match:
        return match.group(1)
    else:
        return city

crime_df = df_crime_concat.copy()

crime_df['City'] = crime_df['City'].apply(clean_crime_city)
crime_df['City'] = crime_df['City'].str.lower().str.title().str.strip()
crime_df['state'] = crime_df['state'].apply(clean_crime_city).str.strip()
crime_df['state'] = crime_df['state'].str.lower().str.title().str.strip()
crime_df['City']

```

```python
crime_df_clean['city'] = crime_df_clean['city'].str.replace(r'St\.', 'Saint', regex=True)
crime_df_clean['city'] = crime_df_clean['city'].str.replace(r'Big Bear', 'Big Bear City', regex=True)
# crime_df_clean['city'] = crime_df_clean['city'].str.replace(r'Waterford Town', 'Waterford', regex=True)
# crime_df_clean['city'] = crime_df_clean['city'].str.replace(r'Town Of East Troy', 'East Troy', regex=True)
# crime_df_clean['city'] = crime_df_clean['city'].str.replace(r'Town Of Madison', 'Madison', regex=True)

def change_town(name):
    a = re.search(r"(.*?) (Town|City|Village|Prairie|Beach)", name)
    b = re.search(r"Town Of (.*)", name)
    if a:
        return a.group(1).strip()
    elif b:
        return b.group(1).strip()
    else:
        return name
crime_df_clean['city_second'] = crime_df_clean['city'].apply(change_town)
    
```
