```python
df_melt = df_melt.rename(columns={'RegionName':'city', 'CountyName':'county', 'State':'state', 'Zori':'rental_price'})
df_clean = df_melt[['city', 'state', 'county', 'Time', 'rental_price']]
```

```python
df_clean.info()
```

```python
df_clean = df_clean.dropna()
```

```python
df_clean.info()
```

```python
df_clean['Time'] = pd.to_datetime(df_clean['Time'])
df_clean['year'] = df_clean['Time'].dt.year
df_clean['month'] = df_clean['Time'].dt.month
```

```python
df_clean.info()
```

```python
us_state_abbreviations = {
    "AL": "Alabama",
    "AK": "Alaska",
    "AZ": "Arizona",
    "AR": "Arkansas",
    "CA": "California",
    "CO": "Colorado",
    "CT": "Connecticut",
    "DE": "Delaware",
    "FL": "Florida",
    "GA": "Georgia",
    "HI": "Hawaii",
    "ID": "Idaho",
    "IL": "Illinois",
    "IN": "Indiana",
    "IA": "Iowa",
    "KS": "Kansas",
    "KY": "Kentucky",
    "LA": "Louisiana",
    "ME": "Maine",
    "MD": "Maryland",
    "MA": "Massachusetts",
    "MI": "Michigan",
    "MN": "Minnesota",
    "MS": "Mississippi",
    "MO": "Missouri",
    "MT": "Montana",
    "NE": "Nebraska",
    "NV": "Nevada",
    "NH": "New Hampshire",
    "NJ": "New Jersey",
    "NM": "New Mexico",
    "NY": "New York",
    "NC": "North Carolina",
    "ND": "North Dakota",
    "OH": "Ohio",
    "OK": "Oklahoma",
    "OR": "Oregon",
    "PA": "Pennsylvania",
    "RI": "Rhode Island",
    "SC": "South Carolina",
    "SD": "South Dakota",
    "TN": "Tennessee",
    "TX": "Texas",
    "UT": "Utah",
    "VT": "Vermont",
    "VA": "Virginia",
    "WA": "Washington",
    "WV": "West Virginia",
    "WI": "Wisconsin",
    "WY": "Wyoming",
    "DC": "District of Columbia"
}

df_clean['state'] = df_clean['state'].apply(lambda x: us_state_abbreviations[x])
df_clean['city'] = df_clean['city'].str.lower().str.title().str.strip()
df_clean['state'] = df_clean['state'].str.lower().str.title().str.strip()
df_clean = df_clean.drop(columns=['county', 'Time'])
```

```python
def change_town(name):
    a = re.search(r"(.*?) (Town|City|Village|Prairie|Beach)", name)
    b = re.search(r"Town Of (.*)", name)
    if b:
        return b.group(1).strip()
    elif a:
        return a.group(1).strip()
    else:
        return name
df_clean['city_two'] = df_clean['city'].apply(change_town)
df_clean.info()
```

```python
city = pd.read_csv('cleaned/city.csv')
city = city[['city_id', 'city', 'state']]
```

```python
get_id = city[['city_id', 'state', 'city']].merge(df_clean, how='right', left_on=['city','state'], right_on=['city', 'state'])
get_id.head()
get_id[get_id['city_id'].isnull()]
```

```python
no_id = get_id[get_id['city_id'].isnull()]
no_id
```

```python
get_id_2 = pd.merge(left=city, right=no_id, how="inner", left_on=['city','state'], right_on=['city_two', 'state'])
get_id_2
```

```python
cols = ['city_id_x', 'city_x', 'state', 'rental_price', 'year', 'month']
get_id_2 = get_id_2[cols]
get_id_2 = get_id_2.rename(columns={
    'city_id_x':'city_id',
    'city_x':'city'
})
get_id_2['city_id'] = get_id_2['city_id'].astype(int)
```

```python
get_id = get_id.dropna(subset=['city_id'])
get_id = get_id.drop(columns=['city_two'])
get_id['city_id'] = get_id['city_id'].astype(int)
```

```python
rent_complete = pd.concat([get_id, get_id_2], ignore_index=True).drop_duplicates(subset=['city_id', 'month', 'year'])
```

```python
rent_complete.info()
```

```python
rent_complete = rent_complete.drop(columns=['state','city'])
rent_complete['city_id'] = rent_complete['city_id'].astype(int)
```

```python
rent_complete.to_csv(os.path.join(os.getcwd(), 'cleaned', 'rent.csv'), index=False)
```
