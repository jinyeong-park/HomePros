```python
# Change types of date columns

df['period_begin'] = pd.to_datetime(df['period_begin'])
df['period_end'] = pd.to_datetime(df['period_end'])
df.info()
```

```python
df['year'] = df['period_begin'].dt.year
df['month'] = df['period_begin'].dt.month
df['year'].unique()
```

```python
# lets look at only a subset of the year from 2016 to 2023
df_select = df[df['year'] > 2015]
df_select.info()
```

```python
columns_to_keep = ['year', 'month', 'city', 'state', 'homes_sold', 'property_type', 'median_ppsf','median_sale_price', 'median_list_price', 'median_list_ppsf', 'pending_sales','new_listings','inventory','price_drops','off_market_in_two_weeks']
df_trimmed = df_select[columns_to_keep]
```

```python
df_trimmed.head()
```

```python
df_trimmed.sort_values(['month', 'year','state','city'], ascending=[True,True,True,True])
```

```python
df_no_all = df_trimmed[df_trimmed['property_type'] != 'All Residential'].reset_index(drop=True)
# df_no_all['city'] = df_no_all['city'].str.replace(r'St\. ', 'Saint ', regex=True)
df_no_all.head()
```

```python
df_no_all.info()
```

```python
def change_town(name):
    name = re.sub(r"St\. ", "Saint ", name.strip())
    a = re.search(r"(.*?) (Town|City|Village|Prairie|Beach)", name)
    b = re.search(r"Town Of (.*)", name)
    if a:
        return a.group(1).strip()
    elif b:
        return b.group(1).strip()
    else:
        return name

df_no_all['city'] = df_no_all['city'].str.lower().str.title().str.strip()
df_no_all['city_2'] = df_no_all['city'].apply(change_town)
df_no_all['state'] = df_no_all['state'].str.lower().str.title().str.strip()
```

```python
city = pd.read_csv('cleaned/city.csv')
city = city[['city_id', 'city', 'state']]
```

```python
merge_1 = pd.merge(left=city, right=df_no_all, how='right', on=['city', 'state'])
merge_1
```

```python
no_id = merge_1[merge_1['city_id'].isnull()]
no_id
```

```python
no_id= no_id.drop(columns=['city_id', 'city'])
```

```python
merge_2= pd.merge(left=city, right=no_id, how="inner", left_on=['city', 'state'], right_on=['city_2', 'state'])
merge_2 = merge_2.drop(columns=['city_2'])
merge_2
```

```python
merge_1 = merge_1.dropna(subset=['city_id'])
merge_1 = merge_1.drop(columns=['city_2'])
```

```python
sales = pd.concat([merge_1, merge_2], ignore_index=True).drop_duplicates(subset=['city_id', 'property_type', 'year', 'month'])
sales['city_id'] = sales['city_id'].astype(int)
```

```python
sales
```

```python
sales.info(verbose=True, show_counts=True)
sales = sales.drop(columns=['city', 'state'])
```

```python
sales.to_csv("cleaned/homesales.csv", index = False)
```
